## 导入相关库

```python
# !/usr/bin/env python
# -*- coding=utf-8 -*-
import pandas as pd
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt
#"inline"可直接在python console里面生成图像
%matplotlib inline

from collections import Counter
from nltk.tokenize import word_tokenize
```

## 数据处理

```python
# 数据加载
data = pd.read_csv('./Market_Basket_Optimisation.csv',header = None)# header = None表示该数据集无header

# 将数据集进行格式转换:逐行构成set,加到tansaction中
words = []
for i in range(0,7500):
    list_temp=data.loc[i].values
    for item in list_temp:
        item=str(item).lower() # 统一小写 
        item=item.strip() # 删除首尾空格换行符
        if item == 'none' or item == 'nan':
            continue
        words.append(item)
# print(words)
```

## 词云分析

```python
# # 去掉停用词
# def remove_stop_words(f):
# 	stop_words = ['Movie']
# 	for stop_word in stop_words:
# 		f = f.replace(stop_word, '')
# 	return f

# 生成词云
def create_word_cloud(f):
	print('根据词频，开始生成词云!')
	# f = remove_stop_words(f)
	cut_text = word_tokenize(f)
	#print(cut_text)
	cut_text = " ".join(cut_text)
	wc = WordCloud(
		max_words=100,
		width=2000,
		height=1200,
    )
	wordcloud = wc.generate(cut_text)
	# 写词云图片
	wordcloud.to_file("wordcloud.jpg")
	# 显示词云文件
	plt.imshow(wordcloud)
	plt.axis("off")
	plt.show()

# 生成词云
words_str=" ".join(words)
create_word_cloud(words_str)
```

```python
# 词频统计
count = Counter(words)
most_count = count.most_common() #list[(key,val),]
print(most_count)

# 词云绘制
wc = WordCloud(
        background_color = 'white',
        max_words=100,
        width=2000,
        height=1200,
    )
wordcloud = wc.generate(str(most_count))
# 写词云图片
# wordcloud.to_file("./wordcloud.jpg")
# 显示词云文件
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
