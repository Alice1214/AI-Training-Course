## 导入相关库

```python
# !/usr/bin/env python
# -*- coding=utf-8 -*-
import pandas as pd
import numpy as np
from efficient_apriori import apriori
```

## 数据处理

```python
# 数据加载
data = pd.read_csv('./Market_Basket_Optimisation.csv',header = None)# header = None表示该数据集无header

# 将数据集进行格式转换:逐行构成set,加到tansaction中
transactions = []
for i in range(0,7500):
    list_temp=data.loc[i].values
    temp_set = set()
    for item in list_temp:
        item=str(item).lower() # 统一小写 
        item=item.strip() # 删除首尾空格换行符
        if item == 'none' or item == 'nan':
            continue
        temp_set.add(item)
    transactions.append(temp_set)
print(transactions)
```

## 关联分析

```python
# 挖掘频繁项集和频繁规则
itemsets, rules = apriori(transactions, min_support=0.02,  min_confidence=0.45)
print('频繁项集：', itemsets)
print('关联规则：', rules)
```
